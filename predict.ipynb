```python
def latest_checkpoint_path():
    path = "lightning_logs"
    folders = os.listdir(path)
    if len(folders) == 0:
        return None
    folders.sort(reverse=True)
    for folder in folders:
        latest_folder = folders[0]
        ckpt_list = glob.glob(f"./{path}/{folder}/checkpoints/*ckpt")
        if len(ckpt_list) > 0:
            return ckpt_list[-1]
    return None
```
